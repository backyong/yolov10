```python
# 1.   
!pip install ultralytics

# 2. YOLOv8  
from ultralytics import YOLO

# 3.  
model = YOLO('/content/yolov8n.pt')  # yolov8n.pt  

# 4.  
import cv2
from matplotlib import pyplot as plt

image_path = '/content/1.jpg'  #    
img = cv2.imread(image_path)

# 5. YOLOv8   
results = model(img)

# 6.    
output_image_path = '/content/output.jpg'

# `results`      
results[0].save(output_image_path)

# 7.   
output_img = cv2.imread(output_image_path)
output_img = cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB)  # OpenCV BGR   RGB 
plt.imshow(output_img)
plt.axis('off')  #  
plt.show()

#    
output_image_path
```

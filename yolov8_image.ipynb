```python
# 1. 필수 라이브러리 설치
!pip install ultralytics

# 2. YOLOv8 모델 로드
from ultralytics import YOLO

# 3. 모델 다운로드
model = YOLO('/content/yolov8n.pt')  # yolov8n.pt 파일 경로

# 4. 이미지 로드
import cv2
from matplotlib import pyplot as plt

image_path = '/content/1.jpg'  # 바탕화면에서 업로드한 이미지 경로
img = cv2.imread(image_path)

# 5. YOLOv8을 사용하여 이미지 인식
results = model(img)

# 6. 인식된 결과 이미지 저장
output_image_path = '/content/output.jpg'

# `results`가 리스트이므로 첫 번째 결과에 접근하여 저장
results[0].save(output_image_path)

# 7. 결과 이미지 출력
output_img = cv2.imread(output_image_path)
output_img = cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB)  # OpenCV에서 BGR로 읽어오기 때문에 RGB로 변환
plt.imshow(output_img)
plt.axis('off')  # 축 제거
plt.show()

# 결과 이미지 링크 제공
output_image_path
```
